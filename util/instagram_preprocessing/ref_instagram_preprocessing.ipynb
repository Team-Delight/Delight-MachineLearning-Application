# 전처리 시작

## 음식 1개인 케이스부터 전처리

### Raw data (food = "김치찌개")

```python
df = pd.read_csv(os.path.join(data_path,extract_csv_lists[0]))
```

```python
df
```

### main_text 모두 합치기

```python
total_text_list = []
for text in df["main_text"]:
    if type(text) == str:
        remove_hash_text = text.replace("#"," ")
        total_text_list.append(remove_hash_text)
    else:
        break
```

```python
total_text_list
```

```python
instagram = Twitter() 
morphs = [] 

for sentence in total_text_list: 
    morphs.append(instagram.pos(sentence))
```

### Konlpy가 분류한 단어 종류

```python
morphs
```

### Instagram을 통해 김치찌개를 가장 잘 나타내는 명사 추출 (해당음식, 맛집 단어 제외)

```python
noun_adj_adv_list=[] 
for sentence in morphs : 
    for word, tag in sentence : 
        if tag in ['Noun'] and (len(word) != 1) and (food_name not in word) and ("맛집" not in word) and ("추천" not in word): 
            noun_adj_adv_list.append(word) 
```

```python
count = Counter(noun_adj_adv_list)
```

```python
list_words = list(count.most_common())
```

### 상위 10개 단어 추출

```python
final_list_words = list_words[:10]
```

```python
final_list_words
```

- 참고 : 추후 분석간 재료 + 인스타 정보간 중복이 발생함에 따라 set자료형 사용 필요

### 데이터프레임 만들기

```python
lists = []
for i in final_list_words:
    lists.append(i[0])
lists.insert(0,food_name)
```

```python
lists
```

```python
record = pd.DataFrame(lists).transpose()
record.columns = ['title']+["feature"+str(num) for num in range(1,11)]
```
