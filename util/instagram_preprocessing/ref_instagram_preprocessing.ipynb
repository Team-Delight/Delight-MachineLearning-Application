### Raw data (food = "ê¹€ì¹˜ì°Œê°œ")

```python
df = pd.read_csv(os.path.join(data_path,extract_csv_lists[0]))
```

```python
df
```

### main_text ëª¨ë‘ í•©ì¹˜ê¸°

```python
total_text_list = []
for text in df["main_text"]:
    if type(text) == str:
        remove_hash_text = text.replace("#"," ")
        total_text_list.append(remove_hash_text)
    else:
        break
```

```python
total_text_list
```

```python
instagram = Twitter() 
morphs = [] 

for sentence in total_text_list: 
    morphs.append(instagram.pos(sentence))
```

### Konlpyê°€ ë¶„ë¥˜í•œ ë‹¨ì–´ ì¢…ë¥˜

```python
morphs
```

### Instagramì„ í†µí•´ ê¹€ì¹˜ì°Œê°œë¥¼ ê°€ì¥ ì˜ ë‚˜íƒ€ë‚´ëŠ” ëª…ì‚¬ ì¶”ì¶œ (í•´ë‹¹ìŒì‹, ë§›ì§‘ ë‹¨ì–´ ì œì™¸)

```python
noun_adj_adv_list=[] 
for sentence in morphs : 
    for word, tag in sentence : 
        if tag in ['Noun'] and (len(word) != 1) and (food_name not in word) and ("ë§›ì§‘" not in word) and ("ì¶”ì²œ" not in word): 
            noun_adj_adv_list.append(word) 
```

```python
count = Counter(noun_adj_adv_list)
```

```python
list_words = list(count.most_common())
```

### ìƒìœ„ 10ê°œ ë‹¨ì–´ ì¶”ì¶œ

```python
final_list_words = list_words[:10]
```

```python
final_list_words
```
